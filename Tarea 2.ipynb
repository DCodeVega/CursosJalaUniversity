```python
class Part():
    def __init__(self, name, attack_level=0, defense_level=0, energy_consumption=0):
        self.name = name
        self.attack_level = attack_level
        self.defense_level = defense_level
        self.energy_consumption = energy_consumption
        self.available = True

    def get_status_dict(self):
        formatted_name = self.name.replace(" ","_").lower()
        print("{}_name:".format(formatted_name), self.name.upper())
        print("{}_status:".format(formatted_name), self.available)
        print("{}_attack:".format(formatted_name), self.attack_level)
        print("{}_defense:".format(formatted_name), self.defense_level)
        print("{}_energy_consump:".format(formatted_name), self.energy_consumption)
        print("-------------------------------------------")

head = Part("Head", attack_level=5, defense_level=10, energy_consumption=5)
head.get_status_dict()

weapon = Part("Weapon", attack_level=15, defense_level=0, energy_consumption=10)
weapon.get_status_dict()

left_arm = Part("Left_arm", attack_level=3, defense_level=20, energy_consumption=10)
left_arm.get_status_dict()

right_arm = Part("Right_arm", attack_level=6, defense_level=20, energy_consumption=10)
right_arm.get_status_dict()

left_leg = Part("Left_leg", attack_level=4, defense_level=20, energy_consumption=15)
left_leg.get_status_dict()

right_leg = Part("Right_leg", attack_level=8, defense_level=20, energy_consumption=15)
right_leg.get_status_dict()
```
